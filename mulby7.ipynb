```python
for i in range(2000,3001):
    int_str=str(i)#converting num to str;int_str=2000
    last_digit=int_str[3]#getting last digit;last_digit=0
    double_int=int(last_digit)*2#double with 2;double_int=0*2=0
    new_int=int_str[0]+int_str[1]+int_str[2]#adding last 3 digits;new_int=200
    new_int1=int(new_int)#convert into int;new_int1=200
    sub=new_int1-double_int#200-0;sub=200
    
    int_str1=str(sub)#converting num into str
    last_digit1=int_str1[2]#getting last digit;last_digit1=0
    double_int1=int(last_digit1)*2#double with 2;0*2=0
    new_int2=int_str1[0]+int_str1[1]#adding last 2 digits;new_int2=20
    new_int2=int(new_int1)
    sub1=new_int2-double_int1#20-0;sub1=20
    if(sub1%7==0):
       print(i,"number is divisible by 7")
    else:
       print(i,"number is not divisible by 7")
    
    

```
